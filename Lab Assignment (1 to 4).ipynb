### Que-6 : Crop central 200 x 200 pixel area and display using matplotlib.

```python
img = cv2.imread('/Users/dhruvpatel/Pictures/istockphoto-610041376-612x612.jpg')
# Show original image

h,w = img.shape[:2] #height x width

# Find center coordinates
center_y = h//2
center_x = w//2

# Define cropping boundaries (100 pixels in each direction)
start_y = center_y - 100
end_y = center_y + 100
start_x = center_x - 100
end_x = center_x + 100

# Crop the region using array slicing
cropped = img[start_y:end_y,start_x:end_x]

plt.figure(figsize=(10,5))

#original image
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

#cropped image
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(cropped,cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')

plt.show()
```

### Que-7 : Rotate image by 180 degrees clockwise using opencv.

```python

# Load image
img = cv2.imread('/Users/dhruvpatel/Pictures/Mr.Bean.jpg')

# Rotate image
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -180, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

# Display both images in ONE row, equal size
plt.figure(figsize=(10,5))

# Original
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

# Rotated
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated by 180")
plt.axis("off")

plt.show()
```

### Que-8 : flip image "car.jpg" left to right using opencv.

```python
img = cv2.imread('/Users/dhruvpatel/Pictures/car.avif')

flipped = img[:,::-1]

# Display both images in ONE row, equal size
plt.figure(figsize=(10,5))

#original image
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

#Filped image
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(flipped,cv2.COLOR_BGR2RGB))
plt.title('Fliped Image')
plt.axis('off')
plt.show()
```

### Que-9 : Convert image from RGB to HSV and display each channel using matplotlib.

```python
img = cv2.imread('/Users/dhruvpatel/Pictures/istockphoto-610041376-612x612.jpg')
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(hsv)
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(H, cmap='hsv')
plt.title('Hue Channel')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(S, cmap='gray')
plt.title('Saturation Channel')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(V, cmap='gray')
plt.title('Value Channel')
plt.axis('off')
plt.show()


plt.imshow(cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)) # not converting back to the original image. but its colors are interpreted using HSVs mapping.
plt.title('HSV Color Space')
plt.axis('off')
plt.show()
```

### Que-10 : Compute MSE between two grayscale images img1 and img2.

```python

img1 = cv2.imread('/Users/dhruvpatel/Pictures/car.avif')
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY),cmap = 'gray')
plt.axis('off')
plt.show()

img2 = cv2.imread('/Users/dhruvpatel/Pictures/car.avif')
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY),cmap = 'gray')
plt.axis('off')
plt.show()


```

```python
from skimage.metrics import mean_squared_error
mse = mean_squared_error(img1,img2)
print(f"MSE:{mse:.2f}")
```
